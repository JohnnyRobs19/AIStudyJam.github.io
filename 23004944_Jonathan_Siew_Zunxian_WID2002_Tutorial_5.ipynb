# Question 1

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the domain for L and K
L = np.linspace(0.1, 300, 100)  # avoid zero to prevent undefined behavior
K = np.linspace(0.1, 300, 100)
L, K = np.meshgrid(L, K)

# Cobb-Douglas production function: P(L, K) = 1.0 * L^0.75 * K^0.25
P = 1.0 * L**0.75 * K**0.25

# Plotting the 3D surface plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(L, K, P, cmap='viridis')

# Adding labels and title
ax.set_title("Cobb-Douglas Production Function", fontsize=14)
ax.set_xlabel("Labour (L)")
ax.set_ylabel("Capital (K)")
ax.set_zlabel("Production Output (P)")

plt.tight_layout()
plt.show()
```

# Question 2

```python
import numpy as np
import matplotlib.pyplot as plt

# Define domain for x and y
x = np.linspace(-10, 10, 200)
y = np.linspace(-10, 10, 200)
x, y = np.meshgrid(x, y)

# Define the function h(x, y) = 4x^2 + y^2
h = 4 * x**2 + y**2

# Plotting the surface plot
fig = plt.figure(figsize=(12, 5))

# Surface plot
ax1 = fig.add_subplot(1, 2, 1, projection='3d')
surf = ax1.plot_surface(x, y, h, cmap='viridis')
ax1.set_title('Surface Plot of $h(x, y) = 4x^2 + y^2$')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('h(x, y)')

# Contour plot
ax2 = fig.add_subplot(1, 2, 2)
contour = ax2.contourf(x, y, h, cmap='viridis')
plt.colorbar(contour, ax=ax2)
ax2.set_title('Contour Map of $h(x, y) = 4x^2 + y^2$')
ax2.set_xlabel('x')
ax2.set_ylabel('y')

plt.tight_layout()
plt.show()
```
